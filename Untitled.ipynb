## Por que não investir nas outras plataformas?

### Instagram

- Embora o Instagram tenha um custo por resultado baixo (0.019 USD), o retorno financeiro (ROAS) é menor que o do Facebook (1.17 vs 1.50).
- A receita total gerada no Instagram é muito menor (54.274 USD) em comparação ao Facebook (248.295 USD), o que limita a escalabilidade do investimento.
- O CPC (Link) é relativamente alto (0.24 USD), indicando que o custo para atrair cliques é maior do que em outras plataformas como Audience Network.

### Audience Network

- Apesar de apresentar o maior ROAS (3.34), o valor absoluto investido e a receita gerada são muito baixos (apenas 1.811 USD), o que indica pouca escalabilidade para um orçamento maior.
- O baixo volume de resultados e receita faz com que essa plataforma não seja adequada para um investimento significativo de 10.000 euros.
- O público e o alcance são muito menores, o que limita o potencial de crescimento do investimento.

---

Portanto, o Facebook se mostra a melhor escolha por oferecer um bom equilíbrio entre custo, retorno e escalabilidade, enquanto as outras plataformas apresentam limitações que tornam o investimento menos eficiente para o orçamento disponível.

```python


df = df.dropna(subset=['Country', 'Amount Spent (USD)', 'Purchase Conversion Value (Facebook Pixel)', 'Purchase (Facebook Pixel)'])

# Agrupar dados por país e calcular métricas
country_stats = df.groupby('Country').agg(
    Total_Investido=('Amount Spent (USD)', 'sum'),
    Total_Retorno=('Purchase Conversion Value (Facebook Pixel)', 'sum'),
    Total_Compras=('Purchase (Facebook Pixel)', 'sum')
).reset_index()


country_stats['CPA'] = country_stats['Total_Investido'] / country_stats['Total_Compras'].replace(0, 1)  # Evitar divisão por zero
country_stats['ROI'] = (country_stats['Total_Retorno'] - country_stats['Total_Investido']) / country_stats['Total_Investido']

# Filtrar países com pelo menos 10 compras (evitar outliers)
country_stats_filtered = country_stats[country_stats['Total_Compras'] >= 10]

# Ordenar por ROI (maior para menor)
country_stats_sorted = country_stats_filtered.sort_values(by='ROI', ascending=False)

# Mostrar resultados
print("Análise por País:")
print(country_stats_sorted[['Country', 'Total_Investido', 'Total_Retorno', 'ROI', 'CPA']].to_string(index=False))

# Selecionar o melhor país
if not country_stats_sorted.empty:
    best_country = country_stats_sorted.iloc[0]
    print("\n=== Melhor País para Investir ===")
    print(f"País: {best_country['Country']}")
    print(f"Total Investido: €{best_country['Total_Investido']:.2f}")
    print(f"Total Retorno: €{best_country['Total_Retorno']:.2f}")
    print(f"ROI: {best_country['ROI']:.2f}x")
    print(f"CPA: €{best_country['CPA']:.2f}")
```
